# Practical 4: Handling missing values and Encoding of categorical data in given dataset.

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("datasets/Data.csv")
df
```

```python
x = df.iloc[:, :-1].values
x
```

```python
df['Age2'] = df['Age'].fillna(df.Age.mean())
df['Salary2'] = df['Salary'].fillna(df.Salary.median())
df
```

```python
im = SimpleImputer(missing_values = np.nan, strategy="mean")
```

```python
df['Age3'] = im.fit_transform(df.Age.values.reshape(-1, 1))
df['Salary3'] = im.fit_transform(df.Salary.values.reshape(-1, 1))
df
```

```python
im = im.fit(x[:,1:])
x[:,1:] = im.transform(x[:,1:])
x
```
